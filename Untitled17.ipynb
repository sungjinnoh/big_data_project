```python
import pandas as pd

# 파일 경로 설정
file_paths = {
    'visitors_count': '/content/방문자수.csv',
    'visitors_by_age_gender': '/content/성·연령별방문자_연령별.csv',
    'regional_visitors_spending': '/content/지역 방문자수_관광지출액 추세.csv',
    'tourist_attractions_ranking': '/content/관광지검색순위.csv',
    'most_visited_regions': '/content/방문자수최다지역.csv',
    'tourism_spending': '/content/관광지출액.csv'
}
# 데이터프레임으로 로딩
dataframes = {}
for name, path in file_paths.items():
    dataframes[name] = pd.read_csv(path, encoding='cp949')  # 한국어 인코딩 적용

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 한글 폰트
plt.rc('font', family='NanumBarunGothic')

df_visitors = dataframes['visitors_count']

plt.figure(figsize=(14, 7))
barplot = sns.barplot(data=df_visitors, x='시도명', y='방문자수')

# Y축 값을 억 단위로 포맷팅
def millions_formatter(x, pos):
    return f'{int(x / 1e8)}(억)'

plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(millions_formatter))

# 바 차트의 각 막대에 데이터 값을 주석
for p in barplot.patches:
    barplot.annotate(f'{p.get_height():,.0f}',
                     (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha='center',
                     va='center',
                     xytext=(0, 10),
                     textcoords='offset points')

plt.xticks(rotation=45)
plt.title('시도별 방문자수')
plt.xlabel('시도명')
plt.ylabel('방문자수')
plt.tight_layout()
plt.show()
```

```python

# 관광지출액 데이터셋 로드
df_tourism_spending_1 = dataframes['tourism_spending']
df_tourism_spending = df_tourism_spending_1.sort_values(by='관광지출액', ascending=True) # 순서대로 나오게 하려고 사용

# '관광지출액' 일반 숫자로 변환
df_tourism_spending['관광지출액'] = df_tourism_spending['관광지출액'].apply(lambda x: float(x))

# 파이그래프 그리기
plt.figure(figsize=(10, 10))
plt.pie(df_tourism_spending['관광지출액'], labels=df_tourism_spending['시도명'], autopct='%1.1f%%', startangle=140)
plt.title('시도별 관광지출액 비율', pad=20) # pad 20은 타이틀이랑 관광지명이 겹처서 20을 줌
plt.axis('equal')
plt.show()
```

```python
import plotly.express as px

df_tourist_attractions_ranking = dataframes['tourist_attractions_ranking']

# '관광지검색순위' 데이터셋에서 '지역명' 별로 검색 건수를 집계
region_counts = df_tourist_attractions_ranking['시도명'].value_counts().reset_index()
region_counts.columns = ['시도명', '검색건수']

# 버블 차트 생성
fig = px.scatter(region_counts, x='시도명', y='검색건수',
                 size='검색건수', color='시도명', hover_name='시도명',
                 title='지역별 관광지 검색 건수',size_max=100)
# 차트 보여주기
fig.show()
```
